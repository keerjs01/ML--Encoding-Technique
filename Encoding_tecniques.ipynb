```python
#Approach 1 --- find and replace 

import pandas as pd

data = pd.read_csv('C:\\Users\\Athiban\\Desktop\\Kee\\Bigg_Boss_India.csv')

data

#here i have a column "language" whcih has (hindi,tamil,malayalam,telugu and kannada), am going to replace with the number

#tamil - 1
#hindi - 2
#malayalam - 3
#telugu - 4
#kannada - 5

cleanup_nums = {"Language": {"Tamil": 1, 
                             "Hindi": 2,
                             "Malayalam": 3,
                             "Telugu": 4,
                             "Kannada": 5
}}

data = data.replace(cleanup_nums)
data.head()
```

```python
#Approach 2 - Label Encoding

data

#gender 
#Female -> 1
#Male -> 2
#Transgender ->2

#1. fiest change the Gender column to category colum
data["Gender"] = data["Gender"].astype('category')
data.dtypes

#cat.codes --> automatically its get converted 

data["gender_new"] = data["Gender"].cat.codes
data.head()

```

```python
#Approach3 - one Hot Encoding

#dummy columns will be created for the values from the columns you give

#note : this since is from pandas library, so we need to use pd.get_dummies
import pandas as pd
pd.get_dummies(data, columns=["Gender","Language"],prefix=["gen","lang"]).head()
```

```python
#Approach4 - Custom Binary Encoding

data["Profession"].value_counts()
```

```python
import numpy as np
data["model_code"] = np.where(data["Profession"].str.contains("Model"),1,0)
data[["Language","Profession","model_code"]].head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
data["prof_code"] = ord_enc.fit_transform(data[["Profession"]])
data[["Profession","prof_code"]].head()
```

```python
from sklearn.preprocessing import OneHotEncoder

oh_enc = OneHotEncoder()
oh_results = oh_enc.fit_transform(data[["Broadcasted By"]])
pd.DataFrame(oh_results.toarray(),columns=oh_enc.categories_).head()
```
